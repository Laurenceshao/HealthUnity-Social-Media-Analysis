```python
# Canada, Sleep Quality, “the number of pos/neg sleep quality tweets in a certain age group” divided by “total number of tweets in this age group in the experiment.” 

ageInfo = df[['output.age.<=18', 'output.age.19-29', 'output.age.30-39', 'output.age.>=40']].reset_index(drop = True)
ageInfo = pd.concat([tmp, ageInfo[:numberOfTweets]], axis = 1)

ageInfo = ageInfo[ageInfo['output.age.<=18'] != 'unknown'] # seems like if there is unknown value then all rows are (i also tried age.19-29... they all yield the same set of rows with "unknown")
# ageInfo[ageInfo['output.age.<=18'] == 'unknown'] # cool, this should be empty now

ageInfo['output.age.<=18'] = pd.to_numeric(ageInfo['output.age.<=18'])
ageInfo['output.age.19-29'] = pd.to_numeric(ageInfo['output.age.19-29'])
ageInfo['output.age.30-39'] = pd.to_numeric(ageInfo['output.age.30-39'])
ageInfo['output.age.>=40'] = pd.to_numeric(ageInfo['output.age.>=40'])



# separate tweets ino the postive and negative groups
posAgeInfo = ageInfo[ageInfo['POS'] > 0.50]
negAgeInfo = ageInfo[ageInfo['NEG'] > 0.50]

# select only the age group columns
posAgeColumns = posAgeInfo.loc[:, ['output.age.<=18', 'output.age.19-29', 'output.age.30-39', 'output.age.>=40']]
negAgeColumns = negAgeInfo.loc[:, ['output.age.<=18', 'output.age.19-29', 'output.age.30-39', 'output.age.>=40']]

# find the index of the maximum value in each row
posIdx = posAgeColumns.idxmax(axis = 1)
negIdx = negAgeColumns.idxmax(axis = 1)

# count the number of different strings in the series
posCounts = posIdx.value_counts()
negCounts = negIdx.value_counts()
totCounts =  posCounts + negCounts

# print(posCounts, totCounts, posCounts / totCounts)

# graph
posAgeBarChart = pd.DataFrame({'Age' : ['<=18', '19-29', '30-39', '>=40'], '% Each Group' : posCounts / totCounts})
posAgeBarChart = posAgeBarChart.plot.bar(x = 'Age', y = '% Each Group', rot = 0, title = 'positive sleep quality tweets divided by age groups (percentage each group)')

negAgeBarChart = pd.DataFrame({'Age' : ['<=18', '19-29', '30-39', '>=40'], '% Each Group' : negCounts / totCounts})
negAgeBarChart = negAgeBarChart.plot.bar(x = 'Age', y = '% Each Group', rot = 0, title = 'negative sleep quality tweets divided by age groups (percentage each group)')
```
