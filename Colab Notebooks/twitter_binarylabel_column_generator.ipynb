```python
with open('/content/drive/Shareddrives/Team HealthUnity/Lpheada-Data-Analysis/passDdtld.pkl' , 'rb') as f:
    data = pickle.load(f)
```

```python
df = data
```

```python
# focus on canada sleep quality tweet now
df = df[df['country'] == 'CA']
df = df[df['lang'] == 'en']
df = df[df['label'] == 'sq']
df.shape
```

```python
type(df['user_id'][55223]) # different from the data type of the id_str belows
```

```python
df['tweet_id'] = df['tweet_id'].astype(str)
df['user_id'] = df['user_id'].astype(str)
df = df.reset_index(drop=True)
df.shape
# type(df['tweet_id'][55223])
```
