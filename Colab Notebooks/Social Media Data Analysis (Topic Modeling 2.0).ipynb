**A deep data cleansing is required for biterm/LDA topic modeling (e.g. delete the stopwords and unnecessary tags).**

```python
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.tokenize import regexp_tokenize

nltk.download('stopwords')
nltk.download('punkt')

# Get English stop words
stop_words = set(stopwords.words("english"))

# Get Spanish stop words
# spanish_stop_words = set(stopwords.words("spanish"))

# Union of both english and spanish
# stop_words = stop_words.union(spanish_stop_words)

# define a function to remove stopwords and tags from a tweet
def process_tweet(tweet):
    # remove website from the tweet
    tweet = re.sub(r'http\S+', '', tweet)
    # remove numbers from the tweet
    tweet = tweet.translate(str.maketrans('', '', string.digits))
    # tokenize the tweet with regular expression
    pattern = r"\w+|[@#]\w+"
    words = regexp_tokenize(tweet, pattern)
    # remove stopwords and tags
    words = [word for word in words if word.lower() not in stop_words and not word.startswith("#") and not word.startswith("@")] 
    """ @ also matters because that information is often irrelvant;
    however #tags might be helpful from my perspective"""
    return " ".join(words)

# apply the function to the tweets column in the dataframe
df["cleaned_tweet"] = df["tweet"].copy()

# upgrade on data cleansing: no numbers, all lower cases
# df['cleaned_tweet'] = df['cleaned_tweet'].str.lower() # consider deleting it
df["cleaned_tweet"] = df["cleaned_tweet"].apply(process_tweet)
# df['cleaned_tweet'] = df['cleaned_tweet'].str.replace('[^a-zA-Z]', ' ') # consider deleting it
```

```python
texts = df['cleaned_tweet'].str.strip().tolist() # canada entire data for now

# PREPROCESSING
# Obtaining terms frequency in a sparse matrix and corpus vocabulary
X, vocabulary, vocab_dict = btm.get_words_freqs(texts)
tf = np.array(X.sum(axis=0)).ravel()
# Vectorizing documents
docs_vec = btm.get_vectorized_docs(texts, vocabulary)
docs_lens = list(map(len, docs_vec))
# Generating biterms
biterms = btm.get_biterms(docs_vec)

# INITIALIZING AND RUNNING MODEL
model = btm.BTM(
    X, vocabulary, seed=12321, T=6, M=2, alpha=0.8642481511542616, beta=0.3023594589411319)
model.fit(biterms, iterations=267)
p_zd = model.transform(docs_vec)

# METRICS
perplexity = btm.perplexity(model.matrix_topics_words_, p_zd, X, 8)
coherence = btm.coherence(model.matrix_topics_words_, X, M=20)
# # or
# perplexity = model.perplexity_
# coherence = model.coherence_

# citation: https://github.com/maximtrp/bitermplus
```

**The following code relates to LDA implementation:**

```python
import re
import pandas as pd
from gensim.corpora import Dictionary
from gensim.models import LdaModel
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('omw-1.4')

# Define regular expression pattern to match URLs
url_pattern = re.compile(r"http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+")

# Define regular expression pattern to match numbers
number_pattern = re.compile(r'\b[0-9]+\b')

# Define a function to remove URLs and numbers from tweets
def remove_urls_numbers(tweet):
    tweet = url_pattern.sub('', tweet) # Remove URLs
    tweet = number_pattern.sub('', tweet) # Remove numbers
    return tweet

# Get English stop words
stop_words = set(stopwords.words("english"))

# Get Spanish stop words
# spanish_stop_words = set(stopwords.words("spanish"))

# Union of both english and spanish
# stop_words = stop_words.union(spanish_stop_words)

# Preprocessing
df['LDA_tweet'] = df['tweet'].str.replace('[^a-zA-Z]', ' ')
df['LDA_tweet'] = df['LDA_tweet'].apply(remove_urls_numbers) # Apply the function to the tweets column in the dataframe
df['LDA_tweet'] = df['LDA_tweet'].str.lower()
df['LDA_tweet'] = df['LDA_tweet'].apply(word_tokenize)
df['LDA_tweet'] = df['LDA_tweet'].apply(lambda x: [word for word in x if word not in stop_words])
df['LDA_tweet'] = df['LDA_tweet'].apply(lambda x: [WordNetLemmatizer().lemmatize(word) for word in x])

# Creating a corpus
dictionary = Dictionary(df['LDA_tweet'])
corpus = [dictionary.doc2bow(text) for text in df['LDA_tweet']]

# Running LDA
num_topics = 10
ldamodel = LdaModel(corpus, num_topics=num_topics, id2word=dictionary)
topics = ldamodel.print_topics()

```
